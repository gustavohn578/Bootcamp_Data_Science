```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

dados = pd.read_csv("/content/A213110189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, 
                    thousands=".", decimal=",", engine='python')

ax = dados.plot(x="Unidade da Federação",y="2008/Ago", kind='bar',figsize=(9,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}")) 

plt.ylabel("Valor total [R$]")
plt.title("Procedimentos hospitalares do SUS - Valor por unidade da federação - Agosto/2008")
plt.show()
```
