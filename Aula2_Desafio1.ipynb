**Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

dados = pd.read_csv("/content/A213524189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, 
                    thousands=".", decimal=",", engine='python')

colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
usaveis = usaveis.drop("Total", axis=1)

ax = usaveis.T.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(loc="upper right", ncol=2, bbox_to_anchor=(1.6, 1))
plt.title("SUS - Valor total de gastos com atendimentos por Unidade da Federação e por ano/mês")
plt.xlabel("Ano/Mês")
plt.ylabel("Valor total [R$]")
plt.grid()
plt.show()
```
