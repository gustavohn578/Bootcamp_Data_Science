Problema: muitas colunas com poucas informacoes. Como trazer somentes as colunas com todas as informações? Varias solucoes possiveis. Ex.:
Media (mean) traz somente resultados para colunas com todos os valores presentes.
Outras estratégias: Fill NA, Drop NA, preencher com media etc

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

mean().index informa colunas com todos os valores presentes

```python
usaveis = dados[colunas_usaveis]
usaveis
```

Problema: index deveria ser a Unidade da Federação
Solução: set_index

```python
usaveis = usaveis.set_index("Unidade da Federação")
usaveis
```
