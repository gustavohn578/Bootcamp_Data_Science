**Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

dados = pd.read_csv("/content/A213524189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, 
                    thousands=".", decimal=",", engine='python')

colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
usaveis = usaveis.drop("Total", axis=1)

ax = usaveis.T[["35 São Paulo","31 Minas Gerais","33 Rio de Janeiro","29 Bahia","41 Paraná"]].plot(figsize=(12,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.legend(loc="upper right")
plt.title("SUS - Valor total de gastos com atendimentos por ano/mês para as 5 unidades da federação mais populosas")
plt.xlabel("Ano/Mês")
plt.ylabel("Valor total [R$]")
plt.grid()
plt.show()
```
