**Desafio: adicionar uma coluna mostrando a região de cada estado.**

---



```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

dados = pd.read_csv("/content/A213524189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, 
                    thousands=".", decimal=",", engine='python')

colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
usaveis = dados[colunas_usaveis]


regiao = [0]*27 #array preenchido com 0
i = 0
while i < 27:
    if usaveis["Unidade da Federação"][i].startswith('1'):
        regiao[i] = "Norte"
    elif usaveis["Unidade da Federação"][i].startswith('2'):
        regiao[i] = "Nordeste"    
    elif usaveis["Unidade da Federação"][i].startswith('3'):
        regiao[i] = "Sudeste"    
    elif usaveis["Unidade da Federação"][i].startswith('4'):
        regiao[i] = "Sul"    
    else:
        regiao[i] = "Centro Oeste"
    i += 1;       

usaveis["Regiao"] = regiao
usaveis = usaveis.set_index("Unidade da Federação")

usaveis["Regiao"]
```
